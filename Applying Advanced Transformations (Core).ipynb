```python
## get the unique ids
unique_ids = sorted(df['Hero'].unique())
unique_ids[:3]
```

```python
## make integers for each id
int_ids = range(len(unique_ids))
int_ids

```

```python
## Converting our range to a list and showing the first 10 values
example_range = list(int_ids)
example_range[:10]
```

```python
# Zip together the unique_ids as the keys and the int_ids as the values
id_map = dict(zip(unique_ids,int_ids))
id_map
```

```python
df['Hero'].map(id_map)
```

```python
## overwriting the original id column 
df['Hero'] = df['Hero'].replace(id_map)
df.head(3)
```

```python
# Using pd.DataFrame and a dictionary
id_lookup = pd.DataFrame({'str_id': id_map.keys(),
                         'int_id':id_map.values()})
id_lookup.head(3)
```

```python
ds.head()
```

```python
## merging with different column names
#pd.merge(ds, df, left_on='hero_names', right_on='hero')

## example merge with our 2 dataframes
merged = pd.merge(ds, id_lookup, left_on='hero_names', right_on='str_id', how='inner')
merged.head(3)



```

```python
merged_df = pd.merge(df, merged, left_on='Hero', right_on='int_id', how='inner')
merged_df.head(3)


```
